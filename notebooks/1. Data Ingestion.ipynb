```python
# Use EDBI G02 Compute server
import pandas as pd
import requests
```

```python
def load_from_s3(file_name):
    """
    Assign to new df df_filename
    """
    try:
        df = (spark.read.format("csv")
                                .option("inferSchema", "true")
                                .option("header", "true")
                                .load(f"s3://edbi-03/s3_teamG02/{file_name}.csv"))
        print(f"{file_name} loaded into df_{file_name}")
        return df.toPandas()
    except:
        print("No file found")
        return
```

```python
# Copy this cell and replace with your own filename and df name
df_industry = load_from_s3("industry")
```

```python
df_poi = load_from_s3("poi")
```

```python
df_employer = load_from_s3("employer")
```

```python
df_case = load_from_s3("case")
```

```python

%sql
CREATE SCHEMA IF NOT EXISTS edbi_teamg02.bronze;
```

```python
# Write the csv into bronze layer data
def write_to_bronze(df, table_name):
    df.write.format("delta").mode("overwrite").option("overwriteSchema","true").option("delta.columnMapping.mode", "name").saveAsTable(f"edbi_teamg02.bronze.{table_name}")
    #log(f"\n edbi_teamg02.bronze.{table_name} written — {df.count()} rows")
    print(f"\n edbi_teamg02.bronze.{table_name} written — {df.count()} rows")
```

```python
# Write the case data into bronze layer data
write_to_bronze(spark.createDataFrame(df_case), "case")
```

```python
# Write the industry data into bronze layer data
write_to_bronze(spark.createDataFrame(df_industry), "industry")
```

```python
# Write the poi data into bronze layer data
write_to_bronze(spark.createDataFrame(df_poi), "poi")
```

```python
# Write the case data into bronze layer data
write_to_bronze(spark.createDataFrame(df_employer), "employer")
```
